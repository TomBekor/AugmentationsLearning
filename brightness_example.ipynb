```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

import torch
import torchvision
import kornia as K
```

```python
def imshow(input: torch.Tensor, size: tuple = None):
    out = torchvision.utils.make_grid(input, nrow=4, padding=5)
    out_np: np.ndarray = K.utils.tensor_to_image(out)
    plt.figure(figsize=size)
    plt.imshow(out_np)
    plt.axis('off')
    plt.show()
```

```python
img_bgr: np.ndarray = cv2.imread('example_images/dog.webp', cv2.IMREAD_COLOR)
x_img: torch.Tensor = K.utils.image_to_tensor(img_bgr)  # CxHxW
x_img = x_img[None,...].float() / 255.
x_img = K.color.bgr_to_rgb(x_img)
```

```python
imshow(x_img)
```

```python
trans_param = .5
```

```python
target_augmentations = torchvision.transforms.Compose([
    K.augmentation.container.ImageSequential(
        K.enhance.AdjustBrightness(brightness_factor=trans_param)
    )
])
```

```python
a_img = target_augmentations(x_img)
```

```python
imshow(torch.cat([x_img, a_img]), size=(10,10))
```
